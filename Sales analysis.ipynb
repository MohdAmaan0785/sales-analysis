```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv" , encoding = "unicode escape") # to avoid encoding error use unicode escape
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Status" , "unnamed1"] , axis = 1 , inplace = True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True) 
# inplace is used to save the changes
```

```python
df["Amount"] = df["Amount"].astype("int")
```

```python
df["Amount"].dtype
```

```python
df.columns
```

```python
df.rename(columns = {"Cust_name":"Customer_Name"})
```

```python
df.describe()
```

```python
df[["Age" , "Orders" , "Amount"]].describe()
```

```python
# Exploratory Data Analysis
```

```python
# Gender
sn = sns.countplot(x = "Gender" , data = df , hue = "Gender")
for bars in sn.containers: # for counting the total amount of males and females
    sn.bar_label(bars)
```

```python
# Total Amount vs Gender
sales_gen = df.groupby(["Gender"] , as_index = False)["Amount"].sum().sort_values(by = "Amount" , ascending = False)
sns.barplot(x = "Gender" , y = "Amount" , data = sales_gen , hue = "Gender" )
```

```python
# From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
```

```python
# Age Group
sn = sns.countplot(x = "Age Group" , data = df , hue = "Gender")
for bars in sn.containers: # for counting the total amount of males and females
    sn.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_gen = df.groupby(["Age Group"] , as_index = False)["Amount"].sum().sort_values(by = "Amount" , ascending = False)
sns.barplot(x = "Age Group" , y = "Amount" , data = sales_gen , hue = "Age Group" )
plt.show()
```

```python
# From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
# Total no. of orders from top 10 states
sales_orders = df.groupby(["State"] , as_index = False)["Orders"].sum().sort_values(by = "Orders" , ascending=False).head(10)
sns.set(rc = {"figure.figsize":(17,5)})
sns.barplot(x = "State" , y = "Orders" , data = sales_orders , hue = "State")
```

```python
# Total Amount/Sales from top 10 States
state_sales = df.groupby(["State"] , as_index = False)["Amount"].sum().sort_values(by = "Amount" , ascending = False).head(10)
sns.set(rc = {"figure.figsize":(17,8)})
sns.barplot(x = "State" , y = "Amount" , data = state_sales , hue = "State")
```

```python
# From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
```

```python
df.columns
```

```python
# Marital Status
sn = sns.countplot(x = "Marital_Status" , data = df , hue = "Gender" )
sns.set(rc={"figure.figsize":(4,5)})
for bars in sn.containers:
    sn.bar_label(bars)
```

```python
# Marital_Status vs Amount
marital_status = df.groupby(["Marital_Status" , "Gender"] , as_index = False)["Amount"].sum().sort_values(by = "Amount" , ascending = False)
marital_status
sns.barplot(x = "Marital_Status" , y = "Amount" , data = marital_status , hue = "Gender")
```

```python
# From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
```

```python
# Occupation
sns.set(rc={"figure.figsize":(20,5)})
sn = sns.countplot(x = "Occupation" , data = df , hue = "Occupation")
for bars in sn.containers:
    sn.bar_label(bars)
```

```python
# Occupation vs Amount
occupation = df.groupby(["Occupation"] , as_index = False)["Amount"].sum().sort_values(by = "Amount" , ascending = False)
sns.barplot(x = "Occupation" , y = "Amount" , data = occupation , hue = "Occupation")
```

```python
# From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector
```

```python
# Product category
sns.set(rc={"figure.figsize":(20,5)})
sn = sns.countplot(x = "Product_Category" , data = df , hue = "Product_Category")
for bars in sn.containers:
    sn.bar_label(bars)
```

```python
# Product Category vs Amount
Product_Category = df.groupby(["Product_Category"] , as_index = False)["Amount"].sum().sort_values(by = "Amount" , ascending = False)
sns.set(rc={"figure.figsize":(30,5)})
sns.barplot(x = "Product_Category", y = "Amount" , data = Product_Category , hue = "Product_Category")
```

```python
# From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
```

```python
# Product_ID vs Orders
pid = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = pid, x = 'Product_ID',y= 'Orders' , hue = "Product_ID")
```
